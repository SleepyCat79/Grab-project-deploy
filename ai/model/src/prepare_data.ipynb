```python
import pandas as pd

df = pd.read_csv('../data/hotel_processed.csv')
df.head()
```

```python
import pandas as pd

df['price'] = df['price'].replace({'VND': '', ',': '', '\xa0': '', '\.': ''}, regex=True)

df['price'] = df['price'].astype(float)

print(df['price'].head())
df
```

```python
df['rating'].fillna(df['rating'].mean(), inplace=True)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Sử dụng PhoBERT
tokenizer = BertTokenizer.from_pretrained("vinai/phobert-base")
model = BertModel.from_pretrained("vinai/phobert-base")

# Hàm chuyển đổi văn bản thành embedding
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    # Lấy embedding của câu (CLS token)
    return outputs.last_hidden_state[:, 0, :].squeeze().numpy()

# Ví dụ: Embedding mô tả khách sạn
df['embedding'] = df['description'].apply(lambda x: get_embeddings(x))
```
