```python
import pandas as pd

df = pd.read_csv('../data/hotel_processed.csv')
df.head()
```

```python
import pandas as pd

df['price'] = df['price'].replace({'VND': '', ',': '', '\xa0': '', '\.': ''}, regex=True)

df['price'] = df['price'].astype(float)

print(df['price'].head())
df
```

```python
df['rating'].fillna(df['rating'].mean(), inplace=True)
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Sử dụng PhoBERT
tokenizer = BertTokenizer.from_pretrained("vinai/phobert-base")
model = BertModel.from_pretrained("vinai/phobert-base")

# Chuyển mô hình và dữ liệu sang GPU (nếu có GPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)


# Hàm chuyển đổi văn bản thành embedding
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding="max_length", truncation=True, max_length=512)
    inputs = {key: value.to(device) for key, value in inputs.items()}  # Chuyển inputs sang GPU nếu có
    with torch.no_grad():
        outputs = model(**inputs)
    # Lấy embedding của câu (CLS token)
    return outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()  # Đảm bảo chuyển về CPU sau khi tính toán

# Chuyển đổi 'room_type' thành chuỗi nếu nó là list
def process_room_type(room_type):
    if isinstance(room_type, list):  # Nếu 'room_type' là list
        return ', '.join(room_type)
    return room_type

# Tạo context cho mỗi khách sạn
df['context'] = df.apply(lambda row: f'''
Đây là mô tả của khách sạn:
{row['description']}
Gía của nó là {row['price']}
Điểm đánh giá của nó là {row['rating']}
Có các loại phòng là {process_room_type(row['room_types'])}
''', axis=1)

# Lấy embedding cho mỗi khách sạn
df['context_embedding'] = df['context'].apply(lambda x: get_embeddings(x))

# Kiểm tra kết quả
print(df[['name', 'context', 'context_embedding']].head())
```
