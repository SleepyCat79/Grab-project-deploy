# Hotel data processing + EDA

## Cleaning

```python
hotel_elderly = dfs['data_hotels_for_elderly_people']
hotel_booking = dfs['hotels_data_from_bookingweb']
```

```python
hotel_elderly.head(3)
```

```python
hotel_booking.head(3)
```

```python
# Add elderly_friendly col to hotel data
hotel_booking['elderly_friendly'] = False
hotel_booking.head(3)
```

```python
# @title Concat data
hotel = pd.concat([hotel_elderly,hotel_booking], ignore_index=True)
hotel.head()
```

```python
hotel['city'] = 'H√† N·ªôi'
```

```python
hotel.info()
```

**Cleaning plan**
- Remove nan (original_price)
- Tax info: no need
- Re-format: rating -> numerical value only
- wdym by facilities? important?
- Re-format: room_types -> H·∫°ng ph√≤ng - gi√° - s·ªë ng∆∞·ªùi

```python
import re

def extract_rating(text):
    if isinstance(text, str):
        match = re.search(r'(\d+,\d+)', text)
        if match:
            return float(match.group(1).replace(',', '.'))
    return None  # or np.nan

hotel['rating'] = hotel['rating'].apply(extract_rating)
```

```python
hotel.head()
```

```python
if hotel['price'].equals(hotel['discounted_price']):
  hotel.drop(columns=['discounted_price'], inplace=True)
hotel.head(3)
```

```python
import numpy as np

hotel.replace('[]',np.nan, inplace = True)

for col in hotel.columns:
  if hotel[col].isnull().values.all(axis=0):
    hotel.drop(columns=[col], inplace=True)

hotel.head(3)
```

```python
hotel.drop_duplicates(inplace=True)
hotel.info()
```
